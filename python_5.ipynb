```python
#simple program in ml
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
years=[[2010],[2011],[2012],[2013],[2014],[2015]]
values=[222,333,444,555,576,666]
train_years,test_years,train_values,test_values=train_test_split(years,values,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(train_years,train_values)
future_year=[[2020]]
predict=model.predict(future_year)
print(f"predicted value for the year {future_year[0][0]}:{predict[0]:.2f}")
```

```python
import pyttsx3
engine=pyttsx3.init()
def speak(text):
    engine.say(text)
    engine.runAndWait()
def main():
    #welcome msg
    speak("Hello! I am your simple bot created by you.")
    speak("you can say hello,ask my name or say goodbye")
    while True:
        command=input("you: ").lower()
    
        if "hello" in command:
            speak("Hi there! welcome to bot world")
        elif "what's your name" or "what is your name" in command:
            speak("hi,My name is simple bot from bot world")
        elif "goodbye" in command:
            speak("goodbye,have a great day")
        else:
            speak("unable to understand say clearly")
if __name__ =="__main__":
    main()
```
