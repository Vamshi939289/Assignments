```python
import re
def clean_text(text):
    text=re.sub(r'RT','',text)
    text=re.sub(r'&amp;','',text)
    text=re.sub(r'[?!.:;,#@-]','',text)
    text=text.lower()
    return text

text="Title: Deep Learning Approaches for Drug Recommendation ,?Authors: Robert King, Linda Scott"
clean=clean_text(text)
print(clean)
```
